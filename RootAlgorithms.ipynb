# Bisection Algorithm

image.png

```
# function = 10-x**2
# function = x**3 - x**2 + 2
function = x ** 3 + 4*x ** 2 - 10
```

```
def bisectionAlgorithm(a, b):
    if function.subs(x, a) * function.subs(x, b) > 0:
        print('Wrong a and b\n')
        return
    elif function.subs(x, a) * function.subs(x, b) == 0:
        if function.subs(x, a) == 0:
            print("Root: ", '%.4f'%a)
        elif function.subs(x, b) == 0:
            print("Root: ", '%.4f'%b)
        else:
            print('An error ocurred')
        return
        
    p = (a+b)/2
    i = 0
    limit = 0

    while (abs(function.subs(x, p)) > 0.000001 and limit < 1000):
        print(str(i) + '\t' +
              '%.32f'%a + '\t' +
              '%.32f'%b + '\t' +
              '%.32f'%p + '\t' +
              '%.32f'%function.subs(x, p))
        if (function.subs(x, p) == 0):
            break
        if (function.subs(x, p)*function.subs(x, a) < 0):
            b = p
        else:
            a = p
            
        p = (a+b)/2
        i += 1
        limit += 1

    print("Root: ", '%.4f' %p)
```

```
bisectionAlgorithm(-2, 2)
```

# False Position Algorithm

image.png

```
function = x ** 3 + 4*x ** 2 - 10
```

```
def falsePositionAlgorithm(a, b):
    if function.subs(x, a) * function.subs(x, b) > 0:
        print('Wrong a and b\n')
        return
    elif function.subs(x, a) * function.subs(x, b) == 0:
        if function.subs(x, a) == 0:
            print('Root: ', '%.4f'%a)
        elif function.subs(x, b) == 0:
            print('Root: ', '%.4f'%b)
        else:
            print('An error ocurred')
        return

    try:
        p = (a * function.subs(x, b) - b * function.subs(x, a))/ (function.subs(x, b) - function.subs(x, a))
    except ZeroDivisionError:
        print("Can't divide by 0")
        return

    limit = 0
    i = 0
    while (abs(function.subs(x, p)) > 0.000001 and limit < 1000):

        print(str(i) + '\t' +
              '%.32f'%a + '\t' +
              '%.32f'%b + '\t' +
              '%.32f'%p + '\t' +
              '%.32f'%function.subs(x, p))

        if function.subs(x, p) == 0: 
            break
        elif function.subs(x, p) * function.subs(x, a) < 0:
            b = p
        else:
            a = p
        try:
            p = (a * function.subs(x, b) - b * function.subs(x, a))/ (function.subs(x, b) - function.subs(x, a))
        except ZeroDivisionError:
            print("Can't divide by 0")
            return
        
        i += 1
        limit += 1
    print("Root: ", '%.4f'%p)
```

```
falsePositionAlgorithm(-2, 2)
```

# Newton's Algorithm

image.png

```
function = x ** 3 + 4*x ** 2 - 10
```

```
def newtonsAlgorithm(root):
    dev = diff(function)

    if function.subs(x, root) == 0:
        print('Root: ', '%.4f'%root)
        return

    p = float(function.subs(x, root)/dev.subs(x, root))

    limit = 0
    i = 0
    while abs(p) > 0.000001 and limit < 1000:

        print('%.f'%limit + '\t' +
              '%.32f'%root + '\t' +
              '%.32f'%function.subs(x, root) + '\t' +
              '%.32f'%dev.subs(x, root))
        
        try:
            p = float(function.subs(x, root)/dev.subs(x, root))
        except ZeroDivisionError:
            print("Can't divide by 0")
            return
        
        root = root - p
        limit += 1
        i += 1
      
    print("Root: ", root)
```

```
newtonsAlgorithm(1)
```

# Secant Algorithm

image.png

```
# function = x ** 5 + x ** 3  + 3
# function = (x-2) ** 2 -4
# function = cos(x)
function = x ** 3 + 4*x ** 2 - 10
```

```
def secantAlgorithm(a, b):
    if function.subs(x, a) * function.subs(x, b) > 0:
        print('Wrong a and b\n')
        return
    elif function.subs(x, a) * function.subs(x, b) == 0:
        if function.subs(x, a) == 0:
            print("Root: ", '%.4f'%a)
        elif function.subs(x, b) == 0:
            print("Root: ", '%.4f'%b)
        else:
            print('An error ocurred')
        return

    try:
        root = float((a * function.subs(x, b) - b * function.subs(x, a)) /
            (function.subs(x, b) - function.subs(x, a)))
    except ZeroDivisionError:
        print("Can't divide by 0")
        return

    i = 0
    limit = 0
    while abs(function.subs(x, root)) > 0.000001 and limit < 1000:

        print(str(i) + '\t' +
            '%.32f'%a + '\t' +
            '%.32f'%b + '\t' +
            '%.32f'%root + '\t' +
            '%.32f'%function.subs(x, root))
        a = b
        b = root

        if (function.subs(x, a) * function.subs(x, root) == 0):
            break

        try:
            h = ((a * function.subs(x, b) - b * function.subs(x, a)) /
                (function.subs(x, b) - function.subs(x, a)))
        except ZeroDivisionError:
            print("Can't divide by 0")
            return

        root = h
        
        i += 1
        limit += 1

    print("Root: ", '%.4f'%root)
```

```
secantAlgorithm(-2, 2)
```
